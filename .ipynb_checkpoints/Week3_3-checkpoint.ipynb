```python
!pip install geocoder

import geocoder # import geocoder

latitude=[]
longitude=[]
for code in final_table['Postal Code']:
    g = geocoder.arcgis('{}, Toronto, Ontario'.format(code))
    print(code, g.latlng)
    while (g.latlng is None):
        g = geocoder.arcgis('{}, Toronto, Ontario'.format(code))
        print(code, g.latlng)
    latlng = g.latlng
    latitude.append(latlng[0])
    longitude.append(latlng[1])
```

```python
len(latitude)
```

```python
len(longitude)
```

```python
se_lat = pd.Series(latitude)
```

```python
se_lon= pd.Series(longitude)
```

```python
final_table['Latitude'] = se_lat.values
final_table['Longitude'] = se_lon.values
```
